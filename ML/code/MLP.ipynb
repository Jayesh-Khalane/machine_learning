```python
# Let's create a simple MLP for XOR using sklearn and visualize the decision boundary using matplotlib

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

# XOR dataset
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 1, 1, 0])

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define and train the MLP
mlp = MLPClassifier(hidden_layer_sizes=(4,), activation='tanh', max_iter=10000, random_state=42)
mlp.fit(X_scaled, y)

# Create a grid to visualize the decision boundary
xx, yy = np.meshgrid(np.linspace(-2, 2, 500), np.linspace(-2, 2, 500))
grid = np.c_[xx.ravel(), yy.ravel()]
grid_scaled = scaler.transform(grid)
preds = mlp.predict(grid_scaled).reshape(xx.shape)

# Plotting
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, preds, alpha=0.4, cmap='coolwarm')
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='coolwarm', edgecolors='k', s=100)
plt.title("MLP Decision Boundary for XOR Problem")
plt.xlabel("Input 1 (scaled)")
plt.ylabel("Input 2 (scaled)")
plt.grid(True)
plt.show()
```
