```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats
import math

x = pd.read_csv('/content/drive/MyDrive/placement.csv', usecols=[0]).squeeze("columns").to_numpy()
x = np.sort(x)
mu = np.mean(x)
variance = np.var(x)
sigma = math.sqrt(variance)

plt.hist(x, bins=20, density=True, alpha=1, color='g')
plt.plot(x, stats.norm.pdf(x, mu, sigma),color='black')
plt.xlabel('CGPA')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Curve')
plt.show()
skewness = stats.skew(x)
kurtosis = stats.kurtosis(x)

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")


if skewness > 0:
    print("The data is right-skewed.")
elif skewness < 0:
    print("The data is left-skewed.")
else:
    print("The data is approximately symmetric.")

if kurtosis > 0:
    print("The data is leptokurtic.")
elif kurtosis < 0:
    print("The data is platykurtic.")
else:
    print("The d23tic.")
```
