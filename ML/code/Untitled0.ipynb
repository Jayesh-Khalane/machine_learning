```python
import re
import requests
import matplotlib.pyplot as plt
import numpy as np

def get_citation_count(identifier, id_type="DOI"):
    """
    Query the Semantic Scholar API for citation count.

    Parameters:
        identifier (str): The DOI or arXiv identifier (without prefix).
        id_type (str): "DOI" or "ARXIV"

    Returns:
        int or None: Citation count if available; otherwise, None.
    """
    base_url = "https://api.semanticscholar.org/graph/v1/paper/"
    if id_type.upper() == "DOI":
        query = f"DOI:{identifier}"
    elif id_type.upper() == "ARXIV":
        query = f"ARXIV:{identifier}"
    else:
        return None

    url = f"{base_url}{query}?fields=citationCount"
    try:
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            return data.get("citationCount", 0)
        else:
            print(f"Failed for {query} (HTTP {response.status_code})")
            return None
    except Exception as e:
        print(f"Exception for {query}: {e}")
        return None

# Replace this multiline string with your actual reference text.
references_text = """
[1]	Ben Mildenhall, Pratul P. Srinivasan, Matthew Tancik, Jonathan T. Barron, Ravi Ramamoorthi, and Ren Ng. 2021. NeRF: representing scenes as neural radiance fields for view synthesis. Commun. ACM 65, 1 (January 2022), 99–106. https://doi.org/10.1145/3503250
[2]	Cui, H.; Gao, X.; Shen, S.; Hu, Z. HSfM: Hybrid structure-from-motion. In Proceedings of the IEEE Conference on Computer Vision and Pattern Recognition, Honolulu, HI, USA, 21–26 July 2017; pp. 1212–1221.
[3]	Y. Rezaei and S. Lee, sat2pc: Estimating Point Cloud of Building Roofs from 2D Satellite Images. 2022. [Online]. Available: https://arxiv.org/abs/2205.12464
[4]	Zhao, Yang & Wu, Shang & Zhang, Jingqun & Li, Sixu & Li, Chaojian & Lin, Yingyan. (2023). Instant-NeRF: Instant On-Device Neural Radiance Field Training via Algorithm-Accelerator Co-Designed Near-Memory Processing. 10.48550/arXiv.2305.05766.
[5]	Schwarz, Katja & Liao, Yiyi & Niemeyer, Michael & Geiger, Andreas. (2020). GRAF: Generative Radiance Fields for 3D-Aware Image Synthesis. 10.48550/arXiv.2007.02442.
[6]	A. Trevithick and B. Yang, GRF: Learning a General Radiance Field for 3D Representation and Rendering. 2021. [Online]. Available: https://arxiv.org/abs/2010.04595
[7]	Yu, Alex & Ye, Vickie & Tancik, Matthew & Kanazawa, Angjoo. (2020). pixelNeRF: Neural Radiance Fields from One or Few Images. 10.48550/arXiv.2012.02190.
[8]	Derksen, Dawa & Izzo, Dario. (2021). Shadow Neural Radiance Fields for Multi-view Satellite Photogrammetry. 10.48550/arXiv.2104.09877.
[9]	Y. Qu and F. Deng, “Sat-Mesh: Learning Neural Implicit Surfaces for Multi-View Satellite Reconstruction,” Remote Sensing, vol. 15, no. 17, p. 4297, Aug. 2023, doi: 10.3390/rs15174297.
[10]	Marí, Roger & Facciolo, Gabriele & Ehret, Thibaud. (2022). Sat-NeRF: Learning Multi-View Satellite Photogrammetry With Transient Objects and Shadow Modeling Using RPC Cameras. 10.48550/arXiv.2203.08896.
[11]	Rakotosaona, Marie-Julie & Manhardt, Fabian & Arroyo, Diego & Niemeyer, Michael & Kundu, Abhijit & Tombari, Federico. (2023). NeRFMeshing: Distilling Neural Radiance Fields into Geometrically-Accurate 3D Meshes. 10.48550/arXiv.2303.09431.
[12]	Tang, Jiaxiang & Zhou, Hang & Xiaokang, Chen & Hu, Tianshu & Ding, Errui & Wang, Jingdong & Zeng, Gang. (2023). Delicate Textured Mesh Recovery from NeRF via Adaptive Surface Refinement. 10.48550/arXiv.2303.02091.
[13]	W. E. Lorensen and H. E. Cline, “Marching cubes,” in ACM eBooks, 1998, pp. 347–353. doi: 10.1145/280811.281026.
[14]	Zhang, Tongtong & Li, Yuanxiang. (2024). psPRF:Pansharpening Planar Neural Radiance Field for Generalized 3D Reconstruction Satellite Imagery. 10.48550/arXiv.2406.15707.
[15]	C. Billouard, D. Derksen, E. Sarrazin, and B. Vallet, “SAT-NGP : Unleashing neural Graphics primitives for fast relightable Transient-Free 3D reconstruction from satellite imagery,” IGARSS 2022 - 2022 IEEE International Geoscience and Remote Sensing Symposium, pp. 8749–8753, Jul. 2024, doi: 10.1109/igarss53475.2024.10641775.
[16]	X. Zhou, Y. Wang, D. Lin, Z. Cao, B. Li, and J. Liu, “SatelliteRF: Accelerating 3D Reconstruction in Multi-View Satellite Images with Efficient Neural Radiance Fields,” Applied Sciences, vol. 14, no. 7, p. 2729, Mar. 2024, doi: 10.3390/app14072729.
[17]	Marí, R., Facciolo, G., & Ehret, T. (2023). Multi-Date Earth Observation NeRF: The Detail Is in the Shadows. In Proceedings of the IEEE/CVF Conference on Computer Vision and Pattern Recognition (CVPR) Workshops (pp. 2035-2045).
[18]	Y. Yao, Z. Luo, S. Li, T. Fang, and L. Quan, MVSNet: Depth Inference for Unstructured Multi-view Stereo. 2018. [Online]. Available: https://arxiv.org/abs/1804.02505
[19]	R. T. Collins, "A space-sweep approach to true multi-image matching," Proceedings CVPR IEEE Computer Society Conference on Computer Vision and Pattern Recognition, San Francisco, CA, USA, 1996, pp. 358-363, doi: 10.1109/CVPR.1996.517097.
[20]	Kendall, Alex & Martirosyan, Hayk & Dasgupta, Saumitro & Henry, Peter & Kennedy, Ryan & Bachrach, Abraham & Bry, Adam. (2017). End-to-End Learning of Geometry and Context for Deep Stereo Regression. 10.48550/arXiv.1703.04309.
[21]	N. Xu, B. Price, S. Cohen and T. Huang, "Deep Image Matting," 2017 IEEE Conference on Computer Vision and Pattern Recognition (CVPR), Honolulu, HI, USA, 2017, pp. 311-320, doi: 10.1109/CVPR.2017.41.
[22]	Yao, Yao & Luo, Zixin & Li, Shiwei & Shen, Tianwei & Fang, Tian & Quan, Long. (2019). Recurrent MVSNet for High-Resolution Multi-View Stereo Depth Inference. 5520-5529. 10.1109/CVPR.2019.00567.
[23]	K. Luo, T. Guan, L. Ju, H. Huang, and Y. Luo, “P-MVSNET: Learning Patch-Wise Matching Confidence Aggregation for Multi-View Stereo,” 2021 IEEE/CVF International Conference on Computer Vision (ICCV), pp. 10451–10460, Oct. 2019, doi: 10.1109/iccv.2019.01055.
[24]	Yang, Jiayu & Mao, Wei & Alvarez, Jose & Liu, Miaomiao. (2019). Cost Volume Pyramid Based Depth Inference for Multi-View Stereo. 10.48550/arXiv.1912.08329.
[25]	Gu, Xiaodong & Fan, Zhiwen & Zhu, Siyu & Dai, Zuozhuo & Tan, Feitong & Tan, Ping. (2019). Cascade Cost Volume for High-Resolution Multi-View Stereo and Stereo Matching. 10.48550/arXiv.1912.06378.
[26]	J. Chen, Z. Yu, L. Ma, and K. Zhang, “Uncertainty awareness with adaptive propagation for multi-view stereo,” Applied Intelligence, vol. 53, no. 21, pp. 26230–26239, Aug. 2023, doi: 10.1007/s10489-023-04910-z.
[27]	J. Yan et al., “Dense Hybrid Recurrent Multi-view Stereo Net with Dynamic Consistency Checking,” in Lecture notes in computer science, 2020, pp. 674–689. doi: 10.1007/978-3-030-58548-8_39.
[28]	X. Shi, Z. Chen, H. Wang, D.-Y. Yeung, kin Wai-Wong, and chun Wang-Woo, Convolutional LSTM Network: A Machine Learning Approach for Precipitation Nowcasting. 2015. [Online]. Available: https://arxiv.org/abs/1506.04214
[29]	Liu, Jin & Ji, Shunping. (2020). A Novel Recurrent Encoder-Decoder Structure for Large-Scale Multi-view Stereo Reconstruction from An Open Aerial Dataset. 10.48550/arXiv.2003.00637.
[30]	Zhang, Jingyang & Yao, Yao & Li, Shiwei & Luo, Zixin & Fang, Tian. (2020). Visibility-aware Multi-view Stereo Network. 10.48550/arXiv.2008.07928.
[31]	Xu, Qingshan & Tao, Wenbing. (2020). PVSNet: Pixelwise Visibility-Aware Multi-View Stereo Network. 10.48550/arXiv.2007.07714.
[32]	Z. Wei, Q. Zhu, C. Min, Y. Chen, and G. Wang, “AA-RMVSNET: Adaptive Aggregation Recurrent Multi-view Stereo Network,” 2021 IEEE/CVF International Conference on Computer Vision (ICCV), Oct. 2021, doi: 10.1109/iccv48922.2021.00613.
[33]	J. Gao, J. Liu, and S. Ji, “A general deep learning based framework for 3D reconstruction from multi-view stereo satellite images,” ISPRS Journal of Photogrammetry and Remote Sensing, vol. 195, pp. 446–461, Dec. 2022, doi: 10.1016/j.isprsjprs.2022.12.012.
[34]	Facciolo, Gabriele & de Franchis, Carlo & Llopis, Eva. (2017). Automatic 3D Reconstruction from Multi-date Satellite Images. 1542-1551. 10.1109/CVPRW.2017.198.
[35]	Lussange, Johann & Yu, Mulin & Tarabalka, Yuliya & Lafarge, Florent. (2023). 3D detection of roof sections from a single satellite image and application to LOD2-building reconstruction. 10.48550/arXiv.2307.05409.
[36]	K. He, G. Gkioxari, P. Dollar, and R. Girshick, “Mask R-CNN,” IEEE International Conference on Computer Vision (ICCV), Oct. 2017, doi: 10.1109/iccv.2017.322.
[37]	R. Girshick, J. Donahue, T. Darrell, and J. Malik, Rich feature hierarchies for accurate object detection and semantic segmentation. 2014. [Online]. Available: https://arxiv.org/abs/1311.2524
[38]	R. Girshick, “Fast R-CNN,” IEEE International Conference on Computer Vision, Dec. 2015, doi: 10.1109/iccv.2015.169.
[39]	S. Ren, K. He, R. Girshick, and J. Sun, Faster R-CNN: Towards Real-Time Object Detection with Region Proposal Networks. 2016. [Online]. Available: https://arxiv.org/abs/1506.01497
[40]	 Yu, D., Ji, S., Liu, J., & Wei, S. (2020). Automatic 3D building reconstruction from multi-view aerial images with deep learning. ISPRS Journal of Photogrammetry and Remote Sensing, 171, 155–170. doi:10.1016/j.isprsjprs.2020.11.011
[41]	Z. Li, Z. Li, Z. Cui, M. Pollefeys, and M. R. Oswald, Sat2Scene: 3D Urban Scene Generation from Satellite Images with Diffusion. 2024. [Online]. Available: https://arxiv.org/abs/2401.10786
[42]	S. Jiang, C. Jiang, and W. Jiang, “Efficient structure from motion for large-scale UAV images: A review and a comparison of SfM tools,” ISPRS Journal of Photogrammetry and Remote Sensing, vol. 167, pp. 230–251, Jul. 2020, doi: 10.1016/j.isprsjprs.2020.04.016.
[43]	Sturm, P.; Triggs, B. A factorization based algorithm for multi-image projective structure and motion. In Proceedings of the Computer Vision—ECCV ’96: 4th European Conference on Computer Vision, Cambridge, UK, 15–18 April 1996; Proceedings Volume II 4; Springer: London, UK, 1996; pp. 709–720.
[44]	B. Curless and M. Levoy, “A volumetric method for building complex models from range images,” in Proceedings of the 23rd Annual Conference on Computer Graphics and Interactive Techniques, 1996, pp. 303–312. doi: 10.1145/237170.237269.
[45]	H. Li, J. Yin, and L. Jiao, “An Improved 3D Reconstruction Method for Satellite Images Based on Generative Adversarial Network Image Enhancement,” Applied Sciences, vol. 14, no. 16, 2024, doi: 10.3390/app14167177.
[46]	Gao, J., Liu, J., & Ji, S. (2021). Rational polynomial camera model warping for deep learning based satellite multi-view stereo matching. In Proceedings of the IEEE/CVF International Conference on Computer Vision (ICCV) (pp. 6148–6157). doi:10.1109/ICCV.2021.6148
[47]	Chen, X.; Diao, W.; Zhang, S.; Wei, Z.; Liu, C. SA-SatMVS: Slope Feature-Aware and Across-Scale Information Integration for Large-Scale Earth Terrain Multi-View Stereo. Remote Sens. 2024, 16, 3474. https://doi.org/10.3390/rs16183474
[48]	Z. Chen, W. Li, Z. Cui and Y. Zhang, "Surface Depth Estimation From Multiview Stereo Satellite Images With Distribution Contrast Network," in IEEE Journal of Selected Topics in Applied Earth Observations and Remote Sensing, vol. 17, pp. 17837-17845, 2024, doi: 10.1109/JSTARS.2024.3457616.
[49]	Zhang, Song & Wei, Zhiwei & Xu, Wenjia & Zhang, Lili & Wang, Yang & Zhang, Jinming & Liu, Junyi. (2025). TS-SatMVSNet: Slope Aware Height Estimation for Large-Scale Earth Terrain Multi-view Stereo. 10.48550/arXiv.2501.01049.
[50]	C. Stucker and K. Schindler, “ResDepth: A deep residual prior for 3D reconstruction from high-resolution satellite images,” ISPRS Journal of Photogrammetry and Remote Sensing, vol. 183, pp. 560–580, Jan. 2022, doi: 10.1016/j.isprsjprs.2021.11.009.
"""

# Extract DOIs using a regular expression.
doi_pattern = re.compile(r'(10\.\d{4,9}/[-._;()/:A-Za-z0-9]+)')
doi_matches = doi_pattern.findall(references_text)

# Extract arXiv IDs from URLs in the text.
arxiv_pattern = re.compile(r'arxiv\.org/abs/(\d+\.\d+)(?:v\d+)?', re.IGNORECASE)
arxiv_matches = arxiv_pattern.findall(references_text)

# Dictionary to hold citation counts by identifier.
citation_data = {}

# Query citation counts for each DOI.
for doi in doi_matches:
    key = f"DOI:{doi}"
    if key not in citation_data:  # avoid duplicates
        count = get_citation_count(doi, "DOI")
        citation_data[key] = count

# Query citation counts for each arXiv ID.
for arxiv_id in arxiv_matches:
    key = f"ARXIV:{arxiv_id}"
    if key not in citation_data:
        count = get_citation_count(arxiv_id, "ARXIV")
        citation_data[key] = count

# Filter out any entries that did not return a count.
filtered_data = {k: v for k, v in citation_data.items() if v is not None}

# Plot the citation counts in a horizontal bar chart.
labels = list(filtered_data.keys())
counts = list(filtered_data.values())

plt.figure(figsize=(12, 8))
plt.barh(np.arange(len(labels)), counts, color='skyblue')
plt.yticks(np.arange(len(labels)), labels)
plt.xlabel("Citation Count")
plt.title("Citation Counts for the References")
plt.tight_layout()
plt.show()
```

```python
import os
from google.colab import drive

def list_files(startpath, output_file):
    with open(output_file, 'w') as f:
        for root, dirs, files in os.walk(startpath):
            level = root.replace(startpath, '').count(os.sep)
            indent = '    ' * level + '|-- '
            f.write(f"{indent}{os.path.basename(root)}/\n")
            subindent = '    ' * (level + 1) + '|-- '
            for file in files:
                f.write(f"{subindent}{file}\n")

def main():
    drive.mount('/content/drive')  # Mount Google Drive
    project_path = '/content/drive/My Drive/Capstone_Project_2024_2025'  # Change this if needed
    output_file = '/content/folder_structure.txt'

    if os.path.exists(project_path):
        list_files(project_path, output_file)
        print(f"Folder structure saved to {output_file}")
    else:
        print(f"Error: Folder '{project_path}' not found. Check the path.")

main()
```
