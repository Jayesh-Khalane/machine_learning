```python
import scipy
import numpy as np

from scipy.constants import pi
#constants
print(scipy.constants.pi)
print(scipy.constants.golden)


#scipy.special
print("factorial of 5 ",scipy.special.factorial(5))

#fft
from scipy.fftpack import fft
x = np.array([1.0, 2.0, 1.0, -1.0, 1.5])
y = fft(x)
print(y)


#transpose
mat=np.matrix("1,3,2;5,6,9;2,5,7")
print(mat.T)


#linspace
print(np.linspace(1,4,9))

#interpolation
from scipy import interpolate
import matplotlib.pyplot as plt
x = np.linspace(0, 4, 25)
y = np.cos(x**2)
f1 = interpolate.interp1d(x, y, kind = 'linear')
xnew = np.linspace(0, 4, 300)
ynew = f1(xnew)
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.legend(['data', 'linear'], loc = 'best')
plt.show()

#3x+2y=2
#x-y=4
#5y+z=-1
a = np.array([[3, 2, 0], [1, -1, 0], [0, 5, 1]])
b = np.array([2, 4, -1])
from numpy import linalg
x = linalg.solve(a,b)
#printing the result array
print("solution of linear eq is {",x,"}")

#cbrt

from scipy.special import cbrt
res = cbrt([10, 9, 0.1254, 27])
print(res)

#log
from scipy.special import logsumexp
import numpy as np
a = np.arange(10)
res = logsumexp(a)
print(res)
```
