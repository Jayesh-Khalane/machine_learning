```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA

# Step 2: Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target  # Not used for training, just for comparison

# Step 3: Apply GMM
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)
labels = gmm.predict(X)

# Step 4: Reduce to 2D for plotting
pca = PCA(n_components=2)
X_2D = pca.fit_transform(X)

# Step 5: Plot GMM clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_2D[:, 0], X_2D[:, 1], c=labels, cmap='viridis', s=50)
plt.title('GMM Clusters on Iris Dataset')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```
