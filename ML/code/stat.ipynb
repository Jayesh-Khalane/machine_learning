```python
def calculate_statistics(data_x, data_y):
    n = len(data_x)

    # Mean
    mean = sum(data_x) / n

    # Median
    sorted_data = sorted(data_x)
    if n % 2 == 0:
        median = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    else:
        median = sorted_data[n // 2]

    # Mode
    frequency = {}
    for num in data_x:
        frequency[num] = frequency.get(num, 0) + 1
    mode = max(frequency, key=frequency.get)

    # Variance
    mean_diff_squared = [(x - mean) ** 2 for x in data_x]
    variance = sum(mean_diff_squared) / n

    # Standard Deviation
    std_dev = variance ** 0.5

    # Covariance and Correlation (if data_y is provided)

    mean_y = sum(data_y) / len(data_y)
    cov_xy = sum((x - mean) * (y - mean_y) for x, y in zip(data_x, data_y)) / n
    std_dev_y = (sum((y - mean_y) ** 2 for y in data_y) / (n - 1)) ** 0.5
    correlation = cov_xy / (std_dev * std_dev_y)

    # Standard Error
    std_error = std_dev / (n ** 0.5)

    # Print results
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}")
    print(f"Variance: {variance}")
    print(f"Standard Deviation: {std_dev}")
    if cov_xy is not None:
        print(f"Covariance: {cov_xy}")
        print(f"Correlation: {correlation}")
    print(f"Standard Error: {std_error}")

# Example data
data_x = [12, 15, 12, 18, 15, 20, 25, 15, 15]
data_y = [22, 25, 20, 30, 28, 35, 40, 32, 30]

calculate_statistics(data_x, data_y)
```
